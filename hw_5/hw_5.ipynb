```python
### 2: KNN
```

```python
data.dropna()
```

```python
data = data.drop(columns = ['Unnamed: 32'])
```

```python
from sklearn import preprocessing

x_data_norm = preprocessing.normalize(data.drop(columns=['diagnosis','id']))
y_data_norm = data['diagnosis'].values
```

```python


X = x_data_norm

y = y_data_norm


```

```python
# 1й вариант обучения
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1, stratify=y)

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)

knn.score(X_test, y_test)
```

```python
# Кросс-валидация

from sklearn.model_selection import GridSearchCV
parameters = {'n_neighbors':[1, 2,3,4, 5, 10, 15, 20, 30, 40, 50, 75, 100]}
knn = KNeighborsClassifier()
clf = GridSearchCV(knn, parameters,cv = 5, scoring = 'accuracy')
clf.fit(X_train, y_train)
print('Best params: ', clf.best_params_)
print('Best accuracy: ', clf.best_score_)
```

```python
При нахождении оптимального значения n_neighbors точность вычисления увеличивается на 2%
```
